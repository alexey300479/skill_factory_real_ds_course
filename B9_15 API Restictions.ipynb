#### Task 1.
Используя API, определите id стотысячного пользователя в группе хабр. Т.е. если бы у нас не было ограничения на размер запроса, то мы бы запросили:
``` python
data['response']['items'][99999]
```

#### Solution:

````python
count = 1000  
offset = 0  
user_ids = []
max_count = data['response']['count']  
while offset < max_count:  
    print(f'Прогресс: {offset / max_count:.2%}', end='\r')     
    params = {  
        'group_id': 'habr',  
        'v': 5.95,  
        'count': count,  
        'offset': offset,  
        'access_token': token  
    }     
    # такой же запрос как в прошлый раз  
    r = requests.get(url, params = params)  
    data = r.json()     
    user_ids += data['response']['items']   
  
    # увеличиваем смещение на количество строк выгрузки  
    offset += count
````
