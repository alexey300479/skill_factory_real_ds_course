```python
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv('tips.csv')
```

```python
df[['total_bill', 'tip']].plot(kind='hist', grid=True, subplots=True, title=['Общая сумма счета', 'Сумма чаевых'], legend=False)
```

```python
df.plot(x='total_bill', y='tip', kind='scatter', grid='True', title='Общая сумма счета VS Сумма чаевых')
```

```python
df.pivot_table(values=['total_bill', 'tip'], index='day', aggfunc='mean').plot(kind='bar')
```

```python
df.pivot_table(values = 'tip', index = ['sex', 'smoker'], aggfunc = 'count').plot(kind = 'bar')
```

```python
df.groupby('smoker')['tip'].mean().plot(kind = 'barh') 
```
