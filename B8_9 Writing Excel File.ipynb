```python
import pandas as pd
```

```python
data = pd.read_excel('nakladnaya.xls', skiprows=2).dropna(how='all')
```

```python
data
```

```python
table = data.iloc[5:7, :].dropna(axis=1, how='all')
```

```python
table
```

```python
table.to_excel('test.xlsx', index=False)
```

```python
writer = pd.ExcelWriter('test.xlsx')
table.to_excel(writer, index=False, sheet_name='Таблица')
writer.save()
```

```python
workbook = writer.book
```

```python
worksheet = writer.sheets['Таблица']
```

```python
money_fmt = workbook.add_format({'bold': True})
name_fmt = workbook.add_format({'color': 'red'})
```

```python
worksheet.set_column('E:F', 20, money_fmt)
worksheet.set_column(1, 1, 20, name_fmt)
```
