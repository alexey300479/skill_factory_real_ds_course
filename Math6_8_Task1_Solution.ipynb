```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
train_df = pd.read_csv('Fram_train.csv')
train_df.drop('Unnamed: 0', axis=1, inplace=True)
train_df.dropna(axis=0, inplace=True)
```

```python
train_df
```

```python
test_df = pd.read_csv('Fram_test.csv')
test_df.drop('Unnamed: 0', axis=1, inplace=True)
test_df.dropna(axis=0, inplace=True)
```

```python
test_df
```

```python
X_train = train_df[['age']]
y_train = train_df[['TenYearCHD']] 
```

```python
X_test = test_df[['age']]
y_test = test_df[['TenYearCHD']] 
```

```python
lm = LogisticRegression(solver='liblinear')
```

```python
model = lm.fit(X_train, y_train.values.ravel()) 
```

```python
model.coef_
```

```python
model.intercept_
```

```python
lm.predict_proba(X_test[:1])
```

```python
lm.score(X_test, y_test)
```

```python
y_pred = lm.predict(X_test)
```

```python
cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
cnf_matrix 
```
