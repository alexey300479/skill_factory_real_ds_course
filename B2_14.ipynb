#### Task 1. С помощью сводной таблицы и функции loc посчитайте, сколько получают ("Wage") русские футболисты ("Russia"), играющие за ФК "AS Monaco".

#### Solution:

```python
pivot = df.pivot_table(values='Wage',
index=['Club'],
columns=['Nationality'],
aggfunc='sum',
fill_value=0)
display(pivot[:5])
```

```python
pivot.loc['AS Monaco']['Russia']
```

#### Cnonclusion: 61000 is the right answer!!!

#### Task 2. Создайте сводную таблицу, содержащую сведения о средней скорости футболистов, занимающих разные позиции (Position) в разных футбольных клубах (Club).

#### Основываясь на данных таблицы, отметьте три позиции, представители которых в среднем обладают самой высокой скоростью.

- RB (правый защитник)
- LM (левый полузащитник)
- RM (правый полузащитник)
- CF (центральный отянутый форвард)
- RWM (правый атакующий полузащитник)
- RS (правый конечный форвард)

#### Solution:

```python
df.columns
```

```python
pivot = df.pivot_table(values = 'SprintSpeed',
                 index=['Position'],
                 columns=['Club'],
                 aggfunc='mean',
                 fill_value=0)
display(pivot[:5])
```

```python
pivot.max(axis=1).sort_values(ascending=False)
```

#### Conclusion: RS, RM, LM are the right answers!!!

#### Task 3. Используя таблицу, созданную на предыдущем шаге, отметьте названия трёх клубов, в которых центральные форварды (ST) обладают наибольшей средней скоростью.

- Sydney FC
- CD Lugo
- Arsenal
- Toronto FC
- Girona FC
- Melbourne Victory
- Vitória

#### Solution:

```python
pivot.loc['ST'].sort_values(ascending=False)[:3].index
```

#### Conclusion: ['Vitória', 'Sydney FC', 'CD Lugo'] are the right answers!!!
