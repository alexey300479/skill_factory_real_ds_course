```python
import json
import pandas as pd
```

```python
with open('recipes.json') as f:
    recipes = json.load(f)
```

```python
ingredients = set()
for recipe in recipes:
    ingredients = ingredients | set(recipe['ingredients'])
len(ingredients)
```

```python
df = pd.DataFrame(recipes)
```

```python
df.head()
```

```python
type(df['ingredients'].iloc[0])
```

```python
def find_item(cell):
    if item in cell:
        return 1
    return 0
```

```python
for item in ingredients:
    df[item] = df['ingredients'].apply(find_item) 
```

```python
df.head()
```

```python
df['ingredients'] = df['ingredients'].apply(lambda x: len(x))
```

```python
df.head()
```

```python
df.to_csv('recipes.csv', index=False)
```
